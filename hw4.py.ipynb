First I imported and investigated into the date file. I indexed and looped through with qualifier statements to weed out the data that didn't have interesting rainfall for June 2015. This code is removed for professional looking notebook clarity.

```python
station_file = pd.read_csv('598354.csv')
station_file.dtypes
```

This station looked pretty good, so I went with it. First I indexed and plotted just the month of June 2015 and its rainfall.

```python
myMonthDates = station_file['DATE'][74528 : 74552] - 20150600
myMonthRain = station_file['PRCP'][74528 : 74552]
```

```python
plot(myMonthDates, myMonthRain, 'b-')
xlabel('Date')
ylabel('RainFall')
title('Rainfall in June 2015 in MANZANITA LAKE CA US')
show()
```

Then I took the entirety of the date given for this particular station and plotted the TMAX and TMIN for it.

```python
myStationYear = range(74602 - 74171)
myStationTMAX = station_file['TMAX'][74171 : 74602]
myStationTMIN = station_file['TMIN'][74171 : 74602]
```

```python
plot(myStationYear, myStationTMAX, 'b-')
xlabel('Days Past Since January 1st, 2014')
ylabel('TMAX')
ylim([0, 400])
title('TMAX in MANZANITA LAKE CA US')
show()
```

```python
plot(myStationYear, myStationTMIN, 'b-')
xlabel('Days Past Since January 1st, 2014')
ylabel('MIN')
ylim([-100, 400])
title('TMIN in MANZANITA LAKE CA US')
show()
```
